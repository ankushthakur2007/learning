```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
sns.relplot(tips,x='total_bill',y='tip',hue='sex',kind='scatter',col='sex')
```

```python
sns.stripplot(tips,x='day',y='total_bill',hue='day',jitter=False)
```

```python
sns.catplot(tips,x='day',y='total_bill',kind='strip',hue='day')
```

```python
sns.catplot(tips,x='day',y='total_bill',kind='swarm',hue='day')
```

```python
sns.catplot(tips,y='total_bill',kind='box')
```

```python
  sns.catplot(tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

```python
sns.kdeplot(tips,x='total_bill',hue='day')
```

```python
sns.catplot(tips,x='sex',y='total_bill',kind='bar')
```

```python
 sns.catplot(tips,x='sex',y='total_bill',kind='point')
```

```python
  sns.catplot(tips,x='sex',kind='count')
```

```python
sns.regplot(tips,x='total_bill',y='tip')
```

```python
 sns.lmplot(tips,x='total_bill',y='tip',hue='sex')
```

```python
 sns.residplot(tips,x='total_bill',y='tip')
```

```python
 g = sns.FacetGrid(tips,col='day',row='time')
 g.map(sns.boxplot,'sex','total_bill')
```

```python
sns.pairplot(iris,hue='species')
```

```python
k= sns.PairGrid(iris,hue='species',vars=['sepal_width','petal_width'])
k.map_upper(sns.scatterplot)
k.map_lower(sns.kdeplot)
k.map_diag(sns.histplot)
k.map(sns.scatterplot)
```

```python
sns.jointplot(tips,x='total_bill',y='tip',kind='scatter')
```
